```python
import torch
```

```python
random_tensor = torch.rand(5)
print(random_tensor)
print(random_tensor.ndim)
```

```python
random_tensor = torch.rand(2,5,3)
print(random_tensor)
print(random_tensor.ndim)
print(random_tensor.shape)
```

```python
torch.manual_seed(42)  # helps to have same random number
random_tensor = torch.rand(5)
print(random_tensor)
```

```python
# generate random numbers integers

random_integers = torch.randint(-10,10, (5,5))
print(random_integers)
```

```python
#generate random num from normal distribution

# torch.normal(mean, std, size)

random_number_normal = torch.normal(0.5, 2, (5,))
print(random_number_normal)
```

```python
# creating noise image using pytorch

import matplotlib.pyplot as plt

# h , w of image 256,256

uniform_noise = torch.rand(255,255)

plt.imshow(uniform_noise.numpy(), cmap='grey')
plt.title("uniform Noise image ")
plt.axis('off')
plt.show()
```

```python
# creating noise image RGB using pytorch

import matplotlib.pyplot as plt

# h , w of image 256x256x3

uniform_noise_rgb = torch.rand(255,255,3)

plt.imshow(uniform_noise_rgb.numpy())
plt.title("uniform Noise image ")
plt.axis('off')
plt.show()
```

```python
print("shape ", uniform_noise_rgb.shape)
print("data typ ", uniform_noise_rgb.dtype)
print("min val ", torch.min(uniform_noise_rgb))
print("max val ", torch.max(uniform_noise_rgb))
```
